```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Metro_Interstate_Traffic_Volume_Cleaned.csv')
data.head()
```

```python
data['date_time'] = pd.to_datetime(data['date_time'])
data['date'] = data['date_time'].dt.date
data['hour'] = data['date_time'].dt.hour
data.head()
```

```python
# expected hours per date
hours_per_day = data.groupby('date')['hour'].nunique()

# count incomplete days
incomplete_days = hours_per_day[hours_per_day < 24]
print(f"Incomplete days: {len(incomplete_days)}")
```

```python
all_dates = pd.date_range(data['date'].min(), data['date'].max())
missing_dates = all_dates.difference(data['date'].unique())
print(f"Missing dates: {len(missing_dates)}")
```
